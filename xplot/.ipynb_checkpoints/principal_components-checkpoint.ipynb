# An nb about clustering PC projected data
## First the PCA

```lua
require 'unsup'
Plot = require 'itorch.Plot'
```

```lua
ns = 1501
ntr = 45

file = torch.DiskFile('../data_loading/test_dat/four_gathers.rsf@', 'r')
file:binary()
raw = file:readFloat(ns*ntr)
file:close()

dat = torch.Tensor(ns,ntr)

for j = 1, ntr do
        for i = 1, ns do
                dat[i][j] = raw[i + (j-1)*ns]
        end
end

dat:add(-dat:mean())
dat:div(dat:std())

vals, vects = unsup.pcacov(dat)

cmprsd = vects[{{},{-2,-1}}]

--new = torch.mm(torch.mm(dat,cmprsd), cmprsd:t())

proj = torch.mm(dat,cmprsd)

print(#proj)
```

```lua
itorch.image(dat:t())
```

```lua
plot = Plot():circle(proj[{{},{1}}]:reshape(1501), proj[{{},{2}}]:reshape(1501), 'red', 'PCs'):draw()
```

## And now for a clustering test

```lua
-- k means
k = 5           -- # of centroids
iter = 100      -- # of iteratiosn
bsz = 10        -- batchsize

centroids, totalcounts = unsup.kmeans(proj, k, iter, bsz, nil, true)
```

```lua
clst = torch.Tensor(ns)
dist = torch.Tensor(k)
-- clustering
        -- calc Euclidean distance to centroids
for i = 1, ns do
        for j = 1, k do
                dist[j] = torch.sqrt((centroids[j] - proj[i]):pow(2):sum())
        end

        -- cluster by minimum distance
        dist, idx = dist:sort()
        clst[i] = idx[1]
end
```

```lua
width = 100
plt = torch.Tensor(width,ns)
for i = 1, width do
    plt[i] = clst
end

itorch.image(plt)
```
